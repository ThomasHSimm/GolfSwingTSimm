```python

import os
import numpy as np
from fastbook import *

cda = 'C:\\Users\\44781\\Documents\\GitHub\\golfdb'
os.chdir(cda)

foldOuta=cda+'\\_dataSel\\'
lsa = os.listdir(foldOuta)

lsa=[folds for folds in lsa if folds[0] =='_']

for folds in lsa:
    if folds == lsa[2]:
        print(folds)
        img_files = get_image_files(foldOuta+folds)

        i=0
        ii=0
        fig,axs=plt.subplots(3,2,dpi=100,figsize=(50,50))
        for fils in img_files:

            im = PILImage.create(fils)
            fnom=str(fils).split('\\')[-1].split('.')[0]
            pts=np.loadtxt(foldOuta+folds+'\\'+fnom+'.txt')

            print(i,ii)
            axs[ii,i].imshow(im)
            marks=['rx','sk','bo','vg','m^']
            [axs[ii,i].plot(pts[jj, 0], pts[jj, 1], mm,markersize=30,markeredgewidth=6) for jj,mm in enumerate(marks)]
            i=i+1
            if i>1:
                i=0
                ii=ii+1
            if ii>2:
                break
        chxs= 'q'#input('Press to continue, q to quit')
        if chxs=='q':
            break
```

```python
import tarfile
import os.path

def make_tarfile(output_filename, source_dir):
    with tarfile.open(output_filename, "w:gz") as tar:
        tar.add(source_dir, arcname=os.path.basename(source_dir))
        
make_tarfile(moveTo+'.tgz',moveTo)
```
